## crangraph

This notebook is used as a scratchpad for testing Python functions used in crangraph, an application designed to provide a streaming view of the R language dependency graph.

```python
# Load Dependencies
import crangraph_utils as cg
import requests as rq
```

To begin our analysis, we need to understand the full universe of packages on CRAN. You can get the current package list using `get_package_list`.

```python
# Grab all packages
pkg_list = cg.get_package_list()
print 'Found {n_pkg} packages on CRAN'.format(n_pkg = len(pkg_list.keys()))

# Print it out (not that big)
pkg_list
```

Use the cell below to control the package that this notebook runs for.

```python
# Global params
PACKAGE_OF_INTEREST = 'lubridate'
```

Let's start by looking at all the versions of this package

```python
# Find versions
pkg_versions = cg.get_old_releases(PACKAGE_OF_INTEREST)
print 'Found {n_ver} releases of {pkg} on CRAN'.format(n_ver = len(pkg_versions.keys()), pkg = PACKAGE_OF_INTEREST)

# Print the entire dictionary of results
pkg_versions
```

Now that we know all the versions of the package, we can go get the metadata for a particular version.

```python
# Grab a DESCRIPTION file
print 'Grabbing the DESCRIPTION file for {pkg} v{ver}...'.format(pkg = PACKAGE_OF_INTEREST, ver = pkg_versions.keys()[0])
desc_url = cg.build_release_path(PACKAGE_OF_INTEREST, pkg_versions.keys()[0])['DESCRIPTION']
desc_text = rq.get(desc_url)
                                 
# parse dependencies
deps = cg.scrape_deps_from_description(desc_text.content)
print 'Dependencies: ' + str(deps)
```
